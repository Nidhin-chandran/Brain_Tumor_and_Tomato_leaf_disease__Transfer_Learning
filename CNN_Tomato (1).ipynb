```python
import pandas as pd
import numpy as np
from skimage.io import imread
from skimage.transform import resize

import os
```

```python
x_train= []
y_train = []

main_path = '/content/drive/MyDrive/tomato/train'

sub_path = os.listdir(main_path)

for Tom in sub_path:

  Tomato = os.path.join(main_path,Tom)

  Tomatos = os.listdir(Tomato)

  for img in Tomatos:

    one_img = os.path.join(Tomato,img)

    img_array = imread(one_img)

    img_resize = resize(img_array,(150,150,1))

    x_train.append(img_resize)

    y_train.append(sub_path.index(Tom))

  print(Tom)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_train,y_train,test_size=0.3,random_state=2)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Input,Conv2D,AvgPool2D,Flatten,MaxPool2D
```

```python
model = Sequential()

model.add(Input(shape=(150,150,1)))
model.add(Conv2D(32,(3,3),activation='relu'))
model.add(MaxPool2D(2,2))
model.add(Flatten())
model.add(Dense(20,activation='relu'))
model.add(Dense(6,activation='softmax'))


```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```
